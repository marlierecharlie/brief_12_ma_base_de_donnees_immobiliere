```python
# importer les csv
val_fonc  = pd.read_csv('Valeurs_foncières.csv')
don_com  = pd.read_csv('donnees_communes.csv')
referentiel_est  = pd.read_csv('fr-esr-referentiel-geographique.csv')
```

```python
# drop duplicates
val_fonc = val_fonc.drop_duplicates()
don_com = don_com.drop_duplicates()
referentiel_est = referentiel_est.drop_duplicates()
```

```python
print(val_fonc.info())
```

```python
# garder les bonnes colonnes
bien=val_fonc[["No voie", "Type de voie", "Voie", 
               "Nombre pieces principales", "Surface reelle bati",
              "Surface Carrez du 1er lot", "Type local", "Code departement", "Code commune"]]

# création de l'id bien
bien.reset_index(inplace = True)
bien = bien.rename(columns = {"index" :'id_bien'})
bien
```

```python
# transformer no voie de float à int
bien["No voie"] = bien["No voie"].where(pd.notnull(bien['No voie']), 0)
bien["No voie"] = bien["No voie"].astype(int)
```

```python
vente = val_fonc[["Valeur fonciere", "Date mutation"]]
```

```python
# valeurs to int
vente["Valeur fonciere"] = vente["Valeur fonciere"].where(pd.notnull(vente["Valeur fonciere"]), 0)
vente["Valeur fonciere"] = vente["Valeur fonciere"].astype(int)
# refaire index
vente.reset_index(inplace = True)
vente = vente.rename(columns = {"index" :'id_vente'})
vente
```
